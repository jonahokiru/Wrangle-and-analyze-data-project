1. The WeRate Dogs twitter archived data (twitter_archived_data.csv), this dataset i manually downloaded it from the provided

link.

2. The second dataset the image_prediction.tsv , on this dataset i downloaded into programatically using the request library. 

3. The tweet_json.text dataset i stored it using twitter API and python tweeps library.
